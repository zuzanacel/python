# BubbleSort

![bubble](bubble.png)


````python
arr = [5, 1, 4, 2, 8]
````

````python
arr = [5, 1, 4, 2, 8]
n = len(arr)
for i in range(n - 1):
    print("0.....", n-i-1)
    for j in range(0, n - i - 1):
        print(arr[j], arr[j + 1])
        if arr[j] > arr[j + 1]:
            arr[j], arr[j + 1] = arr[j + 1], arr[j]
            swapped = True 
    if not swapped:
        break
    else:
        print(arr)
````

# Insertion Sort

![insertion](insertion.png)


````python
arr = [85, 12, 59, 45, 72, 51]
[85]  #sorted 
[12, 59, 45, 72, 51] #unsorted
````

````python
if (n := len(arr)) > 1:
    for i in range(1, n):
        v = arr[i]  # i == 1,  v = arr[1] == 12
        j = i - 1   # j == 0
        while j >= 0 and v < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = v 
        print(arr)
````

````python
arr = [29, 72, 98, 13, 87, 66, 52, 51, 36]
ind = 0
for ind in range(len(arr)):
    min_index = ind  
    for j in range(ind + 1, len(arr)): 
        if arr[j] < arr[min_index]:  # 13 < 29 == > True
            min_index = j  # 3
    arr[ind], arr[min_index] = arr[min_index], arr[ind]
    print(arr)
````
